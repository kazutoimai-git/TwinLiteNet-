```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path1 = "./result/baseline/training_metrics.csv"
file_path2 = "./result/decrease_param/training_metrics.csv"
file_path3 = "./result/add_unet_structure/training_metrics.csv"
```

```python
df1 = pd.read_csv(file_path1)
df2 = pd.read_csv(file_path2)
df3 = pd.read_csv(file_path3)
```

```python
df1
```

```python
df1["epoch"] += 1
df2["epoch"] += 1
df3["epoch"] += 1
```

```python
df1
```

```python
df1 = df1.drop('learning_rate',axis=1)
df2 = df2.drop('learning_rate',axis=1)
df3 = df3.drop('learning_rate',axis=1)
```

```python
df1.columns
```

```python
plt.figure(figsize=(12,7))

plt.plot(df1["epoch"],df1["da_miou"],linestyle="-",label="baseline")
plt.plot(df2["epoch"],df2["da_miou"],linestyle="--",label="decerase_param")
plt.plot(df3["epoch"],df3["da_miou"],linestyle=":",label="add_unet_structure")

plt.title('da_miou') # グラフのタイトル
plt.xlabel('Epoch') # X軸のラベル
plt.ylabel('Score') # Y軸のラベル
plt.legend() # 凡例を表示
plt.grid(False) # グリッド線を表示

plt.savefig("./figure/da_miou")
```

# ll_accのスコア

```python
plt.figure(figsize=(12,7))

plt.plot(df1["epoch"],df1["ll_acc"],linestyle="-",label="baseline")
plt.plot(df2["epoch"],df2["ll_acc"],linestyle="--",label="decerase_param")
plt.plot(df3["epoch"],df3["ll_acc"],linestyle=":",label="add_unet_structure")

plt.title('ll_acc') # グラフのタイトル
plt.xlabel('Epoch') # X軸のラベル
plt.ylabel('Score') # Y軸のラベル
plt.legend() # 凡例を表示
plt.grid(False) # グリッド線を表示

plt.savefig("./figure/ll_acc")
```
